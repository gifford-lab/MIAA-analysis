```python
def normalize(df):
    expts = set([c.split('-')[0] for c in df.columns if 'dpn' in c])
    dpn_cols = list([c for c in df.columns if 'dpn' in c])
    #trim any phrases that have no reads
    norm =df.copy()
    norm = norm[np.all(norm[dpn_cols] != 0,axis=1)]
    for expt in expts:
        norm[expt+'-dpnii'] = norm[expt+'-dpnii']/(np.sum(norm[expt+'-dpnii'])/100000)
        norm[expt+'-dpni'] = norm[expt+'-dpni']/(np.sum(norm[expt+'-dpni'])/100000)
        norm[expt] = norm[expt+'-dpnii']/(norm[expt+'-dpni']+norm[expt+'-dpnii'])
    norm['ed.average'] = (norm['ed1.1']+norm['ed1.2']+norm['ed2.1']+norm['ed2.2'])/4.0
    norm['es.average'] = (norm['es1.1']+norm['es1.2']+norm['es2.1']+norm['es2.2']+norm['es3.1']+norm['es3.2']+norm['es4.1']+norm['es4.2'])/8.0
    return norm

def normalizeP(df,rep):
    expts = set([c.split('-')[0] for c in df.columns if 'dpn' in c])
    dpn_cols = [c for c in df.columns if 'dpn' in c]
    norm =df.copy()
    norm = norm[np.all(norm[dpn_cols] != 0,axis=1)]
    for expt in expts:
        norm[expt+'-dpnii'] = norm[expt+'-dpnii']/(np.sum(norm[expt+'-dpnii'])/100000)
        norm[expt+'-dpni'] = norm[expt+'-dpni']/(np.sum(norm[expt+'-dpni'])/100000)
        norm[expt+'_'+rep] = norm[expt+'-dpnii']/(norm[expt+'-dpni']+norm[expt+'-dpnii'])
    norep_expts = ['.'.join(expt.split('.')[:-1]) for expt in expts]
    for expt in norep_expts:
        norm[expt+'_'+rep+'.average'] = (norm[expt+'.rep1'+'_'+rep]+norm[expt+'.rep2'+'_'+rep])/2.0
    return norm
```

```python
dfP2 = pd.read_table('data/phrases-reads-matrix-ovrexpr-rp2.txt')
dfP1 = pd.read_table('data/phrases-reads-matrix-ovrexpr-rp1.txt')
df = pd.read_table('data/phrases-reads-matrix_24042019.txt')
```

```python
def rename_column(c):
    if 'dpn' in c:
        if '-' in c:
            print(c)
            splitc = c.split('-')
            return splitc[0]+'.'+splitc[2]+'-'+splitc[1]
        elif '.' in c:
            expt = c.split('i.')[1]
            digest = c.split('.')[0]
            return expt+'-'+digest
    else:
        return c
df.columns = [rename_column(c) for c in df.columns]
dfP2.columns = [rename_column(c) for c in dfP2.columns]
dfP1.columns = [rename_column(c) for c in dfP1.columns]
```

```python
for data in [dfP1,dfP2,df]:
    expts = set([c.split('-')[0] for c in data.columns if 'dpn' in c])
    for expt in expts:
        sns.distplot(np.log10(data[expt+'-dpni']+1),label='DpnI')
        sns.distplot(np.log10(data[expt+'-dpnii']+1),label='DpnII')
        plt.xlabel('Log10 Read Counts')
        plt.legend()
        plt.title(expt)
        plt.show()
```

```python
norm_df = normalize(df)
norm_perturb1 = normalizeP(dfP1,'August')
norm_perturb2 = normalizeP(dfP2,'Sept')
```

```python
def get_kmer(d):
    if len(d.split('/')) > 2:
        if len(d.split('/')[2].split(',')) > 2:
            return d.split('/')[2].split(',')[2]
    else:
        return 'None'
def get_type(d):
    if 'S' in d:
        return 'scrambled'
    else:
        return 'motif'
def get_bg(d):
    if len(d.split('/')) > 2:
        if len(d.split('/')[2].split(',')) > 2:
            return '/'.join(d.split('/')[2].split(',')[:2])
    else:
        return 'None'   
replicates=[c for c in norm_df.columns if c[0] == 'e' and 'average' not in c]
norm_perturb = pd.merge(norm_perturb, norm_df[replicates+['es.average','ed.average']], left_index=True, right_index=True)
norm_perturb['kmer'] = [get_kmer(d) for d  in norm_perturb['description']]
norm_perturb['control'] = [get_type(c) for c in norm_perturb['category']]
norm_perturb['background'] = [get_bg(d) for d in norm_perturb['description']]
```

```python
norm_perturb = norm_perturb2
norm_perturb = pd.merge(norm_perturb, norm_df[replicates+['es.average','ed.average']], left_index=True, right_index=True)
cat2desc = {
    "CAT1": "universally opening phrases with GC content between 60 - 70 %",
"CAT2": "universally opening phrases with GC content between 30 - 50 %",
"CAT3": "universally closing phrases with GC content between 60 - 70 %",
"CAT4": "universally closing phrases with GC content between 30 - 50 %",
"CAT5": "opening ES cells using one or more occurrences of one k-mer per phrase",
"CAT6": "opening ES cells using combinations of k-mers per phrase",
"CAT7": "phrases opening ED cells using one or more occurrences of one k-mer per phrase",
"CAT8": "phrases opening ED cells using combinations of k-mers per phrase",
"CAT9": "ES-Salient-TF",
"CAT10": "ES-Salient-Top",
"CAT11": "ES-Native",
"CAT12": "ED-Salient-TF",
"CAT13": "ED-Salient-Top",
"CAT14": "ED-Native",
"CAT15": "SLOT-CNN",
"CAT16": "background"
}
norm_perturb['catdesc'] =[cat2desc[c.strip('S')] for c in norm_perturb['category']]

norm_perturb['kmer'] = [get_kmer(d) for d  in norm_perturb['description']]
norm_perturb['control'] = [get_type(c) for c in norm_perturb['category']]
norm_perturb['background'] = [get_bg(d) for d in norm_perturb['description']]
```

```python
from scipy.stats import mannwhitneyu,ttest_rel,wilcoxon
def rename(d):
    nn = ""
    cell,control = d.split('/')
    nn+=cell
    if control == 'scrambled':
        nn += ' - scrambled'
    return nn

my_pal = {"brachyury": "purple",
          "es":'royalblue',
          'es - scrambled':'lightblue',
          "foxa2": "teal", 
          "brachyury - scrambled":"thistle",
          "foxa2 - scrambled":'c',
         'ed':'red','ed - scrambled':'lightsalmon'}

def plot_kmer(k,expt,expt_name):
    normmelt = pd.melt(norm_perturb,id_vars=['seq','catdesc','control','kmer','background'],
                   value_vars=[expt,'es.average','ed.average'])
    normmelt['varcute'] = [v.split('.')[0] for v in normmelt['variable']]
    normmelt['id'] = normmelt['varcute'] + '/'+normmelt['control']
    keepdiff = normmelt[normmelt['kmer'] == k]
    keepdiff['id_clean'] = [rename(d) for d in keepdiff['id']]
    keepdiff['id_kmer'] = keepdiff['kmer'] + ' ' + keepdiff['id_clean']
    ax = sns.swarmplot(x='id_clean',y='value',data=keepdiff,palette=my_pal,order=['es','es - scrambled',
                                                                                 expt_name,expt_name+' - scrambled',
                                                                                 'ed','ed - scrambled'])
    ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
    plt.xlabel('')
    plt.ylabel('Openness - Dpn signal ratio')
    kmer_phrases = pd.pivot_table(data=norm_perturb[norm_perturb['kmer'] == k],
                                  index=['background','kmer'],columns=['control'],
                                  values=['es.average',expt,'ed.average'])
    kmer_phrases=kmer_phrases.dropna()

    dist1=0.05
    dist2=0.15
    dist3=0.35
    
    expres = wilcoxon(kmer_phrases[expt].motif,kmer_phrases['es.average'].motif)
    expred = wilcoxon(kmer_phrases[expt].motif,kmer_phrases['ed.average'].motif)
    
    scramexpr = wilcoxon(kmer_phrases[expt].motif,kmer_phrases[expt].scrambled)[1]
    scrames = wilcoxon(kmer_phrases['es.average'].motif,kmer_phrases['es.average'].scrambled)[1]
    
    scramed = wilcoxon(kmer_phrases['ed.average'].motif,kmer_phrases['ed.average'].scrambled)[1]
    def ptostar(pval):
        p1 = 'n.s.'
        if pval < 0.05:
            p1 = '*'
        if pval < 0.01:
            p1 = '**'
        if pval < 0.001:
            p1 = '***'
        return p1
    p1 = ptostar(expres[1])
    p2 = ptostar(scrames)
    p3 = ptostar(scramexpr)
    p4 = ptostar(scramed)
    p5 = ptostar(expred[1])
    
    x1,x2 = 2,3
    y, h, col = keepdiff['value'].max() + dist1, dist1, 'k'
    plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
    plt.text((x1+x2)*0.5, y+h, p3, ha='center', va='bottom', color=col)
    x1,x2 = 0,1
    y, h, col = keepdiff['value'].max() + dist1, dist1, 'k'
    plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
    plt.text((x1+x2)*0.5, y+h, p2, ha='center', va='bottom', color=col)
    x1,x2 = 0,2
    y, h, col = keepdiff['value'].max() + dist2, dist1, 'k'
    plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
    plt.text((x1+x2)*0.5, y+h, p1, ha='center', va='bottom', color=col)
 
    x1,x2 = 4,5
    y, h, col = keepdiff['value'].max() + dist1, dist1, 'k'
    plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
    plt.text((x1+x2)*0.5, y+h, p4, ha='center', va='bottom', color=col)
 
    x1,x2 = 2,4
    y, h, col = keepdiff['value'].max() + dist3, dist1, 'k'
    plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
    plt.text((x1+x2)*0.5, y+h, p5, ha='center', va='bottom', color=col)
 
    plt.axis([-0.5,5.5,0,1.4])
    

    plt.title(k)
```

```python
from sklearn.linear_model import LinearRegression
avg_motif=norm_perturb.pivot_table(index=['kmer'],columns=['control'],
                                  values=['es.average','foxa2.24h_Sept-avg','brachyury.24h_Sept-avg','ed.average'])
plt.scatter(y=avg_motif['brachyury.24h_Sept-avg'].motif,x=avg_motif['es.average'].motif,color='k')
lmfit = LinearRegression().fit(np.reshape(np.array(avg_motif['es.average'].motif),(-1,1)),
                               np.reshape(np.array(avg_motif['brachyury.24h_Sept-avg'].motif),(-1,)))
avg_motif['brachyury-es.resid'] = avg_motif['brachyury.24h_Sept-avg'].motif - lmfit.predict(np.reshape(np.array(avg_motif['es.average'].motif),(-1,1)))
top1 = avg_motif.sort_values(by='brachyury-es.resid',ascending=False).head(n=2).index
bottom2 =avg_motif.sort_values(by='brachyury-es.resid',ascending=True).head(n=2).index
interesting_kmers = list(top1)+list(bottom2)
for kmer in list(top1)+list(bottom2):
    y=avg_motif.loc[kmer]['brachyury.24h_Sept-avg'].motif
    x=avg_motif.loc[kmer]['es.average'].motif
    plt.text(x+0.01,y+0.01,kmer)
print(top1)
print(bottom2)
plt.xlabel('stem cell Dpn ratio')
plt.ylabel('brachyury over-expression Dpn ratio')
plt.show()
plt.scatter(y=avg_motif['foxa2.24h_Sept-avg'].motif,x=avg_motif['es.average'].motif,color='k')
plt.xlabel('stem cell Dpn ratio')
plt.ylabel('foxa2 over-expression Dpn ratio')
lmfit = LinearRegression().fit(np.reshape(np.array(avg_motif['es.average'].motif),(-1,1)),
                             np.reshape(np.array(avg_motif['foxa2.24h_Sept-avg'].motif),(-1,)))
avg_motif['foxa2-es.resid'] = avg_motif['foxa2.24h_Sept-avg'].motif - lmfit.predict(np.reshape(np.array(avg_motif['es.average'].motif),(-1,1)))
top5 = avg_motif.sort_values(by='foxa2-es.resid',ascending=False).head(n=2).index
bottom5 =avg_motif.sort_values(by='foxa2-es.resid',ascending=True).head(n=2).index
interesting_kmers.extend(list(top5)+list(bottom5))
for kmer in list(top5)+list(bottom5):
    y=avg_motif.loc[kmer]['foxa2.24h_Sept-avg'].motif
    x=avg_motif.loc[kmer]['es.average'].motif
    if x < 0.35 and y < 0.4:
        plt.text(x+0.01,y+0.018,kmer)
    else:
        plt.text(x+0.01,y,kmer)
plt.show()
```

```python
for kmer in list(top5)+list(bottom5):
    plot_kmer(kmer,'foxa2.24h_Sept-avg','foxa2')
    plt.show()
for kmer in list(top1)+list(bottom2):
    plot_kmer(kmer,'brachyury.24h_Sept-avg','brachyury')
    plt.show()
```
